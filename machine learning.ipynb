```python
from scipy import ndimage
```

```python
bacter=plt.imread('bacteria.png')
bacter=bacter[:,:,0]
plt.imshow(bacter)
bacter.shape
```

```python
bacter2=np.copy(bacter)
plt.hist(bacter2.ravel(),bins=255)
plt.show()
```

```python
bacter2=np.copy(bacter)
bacter2=bacter2<0.6
plt.imshow(bacter2)
plt.show()
```

```python
# morphologie utilisée pour enlever les artefacts
open_bac = ndimage.binary_opening(bacter2)
plt.imshow(open_bac)
```

```python
# Segmentation de l'image: label_image contient les différents labels et n_labels est le nombre de labels
label_image, n_labels = ndimage.label(open_bac)
print(f'il y a {n_labels} groupes')
```

```python
sizes = ndimage.sum(open_bac, label_image, range(n_labels))
```

```python
plt.scatter(range(n_labels), sizes)
plt.xlabel('bactérie ID')
plt.ylabel('taille relative')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel('titanic3.xls')
```

```python
data.loc[data['age']<=20,'age']=0
data.loc[(data['age']>20) & (data['age']<=30),'age']=1
data.loc[(data['age']>30) & (data['age']<=40),'age']=2
data.loc[data['age']>40,'age']=3
```

```python
def cat(age) :
    if age<=20 :
        return 0
    if 20<age<=30 :
        return 1
    if 30<age<=40 :
        return 2
    if age>40:
        return 3
```

```python
data['age']=data['age'].map(cat)
```

```python
data['age'].value_counts()
```

# ANALYSE DE DONNES FINANCE BITCOIN

```python
btc=pd.read_csv('BTC-EUR.csv',index_col='Date',parse_dates=True)
```

```python
data=btc.copy()
```

```python
data['buy']=np.zeros(len(data))
data['sell']=np.zeros(len(data))
data['rolingmax']=btc['Close'].shift(1).rolling(window=28).max()
data['rolingmin']=btc['Close'].shift(1).rolling(window=28).min()
data.loc[data['rolingmax']<data['Close'],'buy']=1
data.loc[data['rolingmin']>data['Close'],'sell']=-1
```

```python
plt.figure(figsize=(12,8))
data['2019']['Close'].plot()
data['2019']['rolingmax'].plot()
data['2019']['rolingmin'].plot()
```

```python
plt.figure(figsize=(12,8))
data['2019']['buy'].plot()
data['2019']['sell'].plot()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

```

```python
data=pd.read_excel('titanic3.xls')
data.columns
```

```python
data=data.drop(['sibsp', 'parch', 'ticket',
       'fare', 'cabin', 'embarked', 'boat', 'body', 'home.dest', 'name'],axis=1)
data.columns
```

```python
data=data.dropna(axis=0)
data.columns
```

```python
data.describe()
```

```python
data.head()
data.columns
```

```python
X=data[['pclass', 'sex', 'age']]
Y=data['survived']
X.loc[X['sex']=='female','sex']=0
X.loc[X['sex']=='male','sex']=1
```

```python
Xtrain,Xtest,Ytrain,Xtest=train_test_split(X,Y,test_size=0.2)
```

```python
plt.hist(data['pclass'],bins=3)
plt.show()
```

```python
model=KNeighborsClassifier(n_neighbors=1)
```

```python
model.fit(Xtrain,Ytrain)
```

```python
model.score(Xtest,Ytest)
```

```python

pre=np.array([1,1,20]).reshape(1,3)
model.predict(pre)
```
